```python
import pandas as pd
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_features.csv')
```

```python
#Let's see what our distribution of workaholism looks like
df[['Workaholism']].plot(kind = 'hist')
```

```python
labels = '1','2','3','4','5'


one = len(df[df['Workaholism'] == 1])
two = len(df[df['Workaholism'] == 2])
three = len(df[df['Workaholism'] == 3])
four = len(df[df['Workaholism'] == 4])
five = len(df[df['Workaholism'] == 5])


sizes = [one,two,three,four,five]
colors = ['yellowgreen','lightskyblue', 'gold',  'lightcoral', 'red']

# Plot
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Workaholism")
plt.axis('equal')



```

```python
#Now let's try to find correlations
sns.jointplot(df['Prioritising workload'], df['Workaholism'], kind = 'kde')
```

We see that most values trend towards 3 in both.
We also see that people who don't prioritize workload tend to be workaholic

```python
sns.jointplot(df['Healthy eating'], df['Workaholism'], kind = 'kde')
```

```python
sns.jointplot(df['Classical music'], df['Workaholism'], kind = 'kde')
```

Summary: for the most part all the distributions seem to be around the same with a couple of differences here and there when it comes to thinking ahead, weight, and classical music.
