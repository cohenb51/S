```python
import pandas as pd
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read in data
df = pd.read_csv('responses.csv')
df.shape
```

```python
#This is our target parameter
phobia = df[['Public speaking']]
phobia.isna().sum()
```

```python
#Some music features
music = df[['Music', 'Classical music', 'Rock', 'Opera', 'Swing, Jazz']]
```

```python
music.head()
```

```python
#Let's see how many nulls we have
music.isna().sum()
```

```python
music[music.isna().any(axis = 1)]
```

```python
#We don't have so many nulls so let's fill with the avg
music = music.fillna(music.mean())
music.loc[643]
```

```python
#Let's repeat for movies. 
movies = df[['Movies', 'Thriller', 'Comedy', 'Documentary', 'Fantasy/Fairy tales', 'Action']]
```

```python
movies.head()
```

```python
movies.isna().sum()
```

```python
movies = movies.fillna(movies.mean())
```

```python
#Let's repeat the process with hobbies
hobbies = df[['Art exhibitions', 'Science and technology', 'Religion', 'Countryside, outdoors', 'Fun with friends','Dancing', 'Law', 'Musical instruments', 'Prioritising workload','Writing notes', 'Funniness', 'Fake', 'Decision making', 'Empathy', 'Judgment calls','Loneliness','Number of friends','Appearence and gestures']]
```

```python
hobbies.head()
```

```python
hobbies.isna().sum()
```

```python
hobbies = hobbies.fillna(3)
hobbies.isna().sum()

```

```python
#Let's repeat with lifeviews
life_views = df[['Personality', 'Assertiveness', 'Workaholism', 'Healthy eating', 'Thinking ahead']]
```

```python
life_views.head()
```

```python
life_views.isna().sum()
```

```python
life_views = life_views.fillna(life_views.mean())
```

```python
spending = df[['Finances', 'Spending on looks']]
```

```python
spending.head()
```

```python
spending.isna().sum()
```

```python
spending = spending.fillna(spending.mean())
spending.head()
```

```python
demographics = df[['Age', 'Weight', 'Number of siblings', 'Education']]
```

```python
demographics.isna().sum()
```

```python
demographics = demographics.fillna(demographics.mean())
```

```python
demographics.head()
```

```python

#Make it easier to add features later
adds = df[['Spending on gadgets', 'Height']]
adds= adds.fillna(adds.mean())
```

```python
features = pd.concat([phobia, movies, music,hobbies, life_views, spending, demographics, adds], axis = 1 )
```

```python
features.shape
```

```python
features.fillna(features.mean())
```

```python
#Let's make sure our types are all numerical
df['Education'].unique()
degrees = {
    'college/bachelor degree' :1,
    'secondary school': 2,
    'primary school' : 3,
    'masters degree' : 4,
    'currently a primary school pupil': 3,
    'doctorate degree': 5,
}
features['Education'] = df['Education'].map(degrees)
features['Education'] = features['Education'].fillna(features['Education'].mean())
features['Happiness in life'] = df['Happiness in life'].dropna()


features2 = df.fillna(df.mean())
df['Alcohol']

```
